## Code Contexts

Code in the `/packages/syft` directory as part of the `syft` pypi package runs in a variety of places. This is because the same code needed for a user to connect and send messages to a server, is the same code that the server needs to decode those messages and reply or to connect to another server and do the same.

Code in the `/packages/grid` directory is almost exclusively used to build and run node servers like a `domain` or `network`.

The important thing to note here is that the `domain` must be reachable to `client` but the client is not able to be reached by the domain except by replying to an incoming request. This is a common pattern on the web. Web servers usually cannot initiate a connection to your home computers browser out of the blue.

The code running inside the client is on the Users computer, in a python interpreter, most commonly Jupyter Lab (like where we are now).

### Importing Syft

First lets import `syft` so we can use it to create a client to our domain like the diagram above.

````python
# ‚öîÔ∏è Runnable Code
import syft as sy
````

Now we can create a client by using the handy `login` function. The return value of this function is a client object which we will name `domain_client`.

### Creating a Client

````python
# ‚öîÔ∏è Runnable Code
domain_client = sy.login(port=8081, email="info@openmined.org", password="changethis")
````

Ultimately inside the client you will find a few important things.  
The name of the node it is pointing to:

````python
# ‚öîÔ∏è Runnable Code
domain_client.name
````

This client points to a domain called `test_domain` which matches the name we used during the `hagrid launch` command above.

````python
# ‚öîÔ∏è Runnable Code
domain_client.address
````

It has an address which is a UUIDv4, this is randomly generated and essentially guaranteed to be universially unique.

<blockquote><strong>üîó Web Link</strong><br /><a href="https://en.wikipedia.org/wiki/Universally_unique_identifier#Version_4_(random)">https://en.wikipedia.org/wiki/Universally_unique_identifier</a></blockquote>

````python
# ‚öîÔ∏è Runnable Code
domain_client.address.target_id
````

And finally we have at least 1 `route` to contact it.

````python
# ‚öîÔ∏è Runnable Code
domain_client.routes[0].connection.base_url
````

This matches the fact that we connected to it over port `8081` on the `localhost` interface which is default when starting a server on a local Desktop setup.

Lets see where the `private key` or `signing_key` which is used to sign all messages is kept. This key is just bytes and we often represent it as `hex` to make display or storage easier.

````python
# ‚öîÔ∏è Runnable Code
print(type(domain_client.signing_key))
bytes(domain_client.signing_key).hex()
````

We logged in as the root (Administrator) using the default credentials so this `signing_key` is quite powerful, so its best not to post these online or print them in your notebooks by accident.

We have a container service called `proxy` which runs a system called `traefik` which allows us to define different HTTP `routes` or `paths` for our client to access.

The first thing to know is that our Message protocol is based on `RPC` not `REST` so most of the application is behind one or two route endpoints, in particular the `/api/v1/syft` and `/api/v1/syft/stream`.

We will go into the details about the difference between those two in the next session.

## Docker Compose

To understand how this works lets look at our `docker-compose.yml` file.

```yaml
proxy:
  restart: always
  hostname: ${DOMAIN_NAME?Variable not set}
  image: ${DOCKER_IMAGE_TRAEFIK?Variable not set}:${TRAEFIK_VERSION?Variable not set}
  networks:
    - "${TRAEFIK_PUBLIC_NETWORK?Variable not set}"
    - default
  volumes:
    - "./traefik/docker/traefik.yml:/etc/traefik/traefik.yml"
    - "./traefik/docker/dynamic.yml:/etc/traefik/conf/dynamic.yml"
  environment:
    - SERVICE_NAME=proxy
    - RELEASE=${RELEASE:-production}
    - HOSTNAME=${DOMAIN_NAME?Variable not set}
    - HTTP_PORT=${HTTP_PORT}
    - HTTPS_PORT=${HTTPS_PORT}
  ports:
    - "${HTTP_PORT}:81"
  depends_on:
    - "docker-host"
```

What you can see here is we use a lot of `env` variables which either come from the default set in our `.env` file or they are passed in during launch by `hagrid`.

The defaults for the above variables are likely to be:
```bash
DOMAIN_NAME=test_domain
DOCKER_IMAGE_TRAEFIK=traefik
TRAEFIK_VERSION=v2.8.1
RELEASE=production
HOSTNAME=test_domain
HTTP_PORT=8081
HTTPS_PORT=443
```

We can also see that some configuration is passed into `traefik` using those volume mounts.

```yaml
# /packages/grid/traefik/docker/dynamic.yml
services:
  backend:
    loadBalancer:
      servers:m
        - url: "http://proxy:8001"
  backend-stream:
    loadBalancer:
      servers:
        - url: "http://proxy:8011"

routers:
  backend:
    rule: "PathPrefix(`/api`)
    entryPoints:
      - web
      - vpn
    service: "backend"
  backend-stream:
    rule: "PathPrefix(`/api`) && PathPrefix(`/api/v1/syft/stream`)
    entryPoints:
      - web
      - vpn
    service: "backend-stream"
```

To aid in understanding unrelated code has been removed from the above.

You can see here that if you access the main port we are listening on, in this case `8081` that any traffic to `/api` will go to the backend container, and any traffic to `/api/v1/syft/stream` will go to the `backend_stream` container.

You will see that these match up to the diagram above, and show the flow of input into two containers which both have a `Node Object` and what looks like a copy of `syft`.

### Backend

Let's take a look at the `backend.dockerfile`.

```bash
# packages/grid/backend/backend.dockerfile

# copy grid
COPY grid/backend /app/

COPY grid/backend/docker-scripts/start.sh /start.sh
COPY grid/backend/docker-scripts/start-reload.sh /start-reload.sh
COPY grid/backend/worker-start.sh /worker-start.sh
COPY grid/backend/worker-start-reload.sh /worker-start-reload.sh

RUN chmod +x /start.sh
RUN chmod +x /start-reload.sh
RUN chmod +x /worker-start.sh
RUN chmod +x /worker-start-reload.sh

# install syft
RUN --mount=type=cache,target=/root/.cache \
  pip install --user -e /app/syft

# change to worker-start.sh or start-reload.sh as needed
CMD ["bash", "start.sh"]
```

Here we can see that essentially, the `backend` container is essentially a bunch of files from the `packages/grid/backend` folder and some scripts for startup.

The other main thing other than lots of platform specific dependencies is we install a copy of `syft`.

It turns out we use this same image for 3 different containers:
- `backend`
- `backend_stream`
- `celeryworker`

Let's take a look at our `docker-compose.yml` definition.

```yaml
backend:
  restart: always
  image: "${DOCKER_IMAGE_BACKEND?Variable not set}:${VERSION-latest}"
  depends_on:
    - proxy
    - db
  env_file:
    - .env
  environment:
    - SERVICE_NAME=backend
    - RELEASE=${RELEASE:-production}
    - VERSION=${VERSION}
    - VERSION_HASH=${VERSION_HASH}
    - NODE_TYPE=${NODE_TYPE?Variable not set}
    - DOMAIN_NAME=${DOMAIN_NAME?Variable not set}
    - STACK_API_KEY=${STACK_API_KEY}
    - PORT=8001
    - IGNORE_TLS_ERRORS=${IGNORE_TLS_ERRORS?False}
    - HTTP_PORT=${HTTP_PORT}
    - HTTPS_PORT=${HTTPS_PORT}
    - USE_BLOB_STORAGE=${USE_BLOB_STORAGE}
    - CONTAINER_HOST=${CONTAINER_HOST}
  command: /start.sh
  network_mode: service:proxy
```

The primary difference between all three definitions are the port which is different between `backend` and `backend_stream`, that is because when we use `network_mode` we share a network device between multiple containers; and the `command` which is different for the `celeryworker` which also does not have a `port` because it is not directly reachable.

The ports and `network_mode` relates to using our bi-directional `vpn` without additional complex routing, which we will discuss later.

### Sync vs Async

The reason we have three containers is to provide two different mechanisms to send in commands to a node. In the next session we will discuss this in more detail but messages which expect an immediate response are considered `sync` and blocking, examples are `login` or `get`.

Examples of `async` messages are most `actions` which are sent to the `stream` endpoint, `queued` and then `consumed` by our `worker` and processed. In this instance there is no reply to `async` messages except an acknowledgement (`ack`) that they were queued.

The reason we have random ports is because it can be useful to start multiple nodes on your machine to test things like connecting a `domain` to a `network` in which case we can't predict which ports will be available.

The `dev_tool.sh` takes a `port` as its first argument, and attempts to open that resource in the relevant tool. In the case of `db`, `redis`, and `mongodb` (in 0.8) it will launch `TablePlus`. If you ask for the ports for `seaweedfs` or `queue` (rabbitmq) it will open up our browser to their built in web management UIs.

The default username and password for rabbitmq is `guest` and `guest`. It is normal practise to only include the UI in development mode which is exactly what the `docker-compose.dev.yml` does.

We will see these in action below once we look at the different stores.

### Weird Emojis

You might notice where it says <UID: ‚ç∞üöÉ> there are two little emojis. ### Stream & Worker

Okay so what about this `backend_stream` thing and the `celeryworker`? The `stream` endpoint and container takes the messages and queues them into our `rabbitmq` queue where they sit waiting to be consumed by our pool of workers. In this case we have a single worker container `test_domain-celeryworker-1`, but you can imagine this can be extended.

Let's look at the container map again. Here we can see that messages coming in on the `stream` endpoint do indeed pass through the `backend_stream` container and go to the `rabbitmq` container next, before making their way to the `celeryworker` container to be processed.

Simply put, the work that an `async` action takes will be queued and executed on the `celeryworker` and will not cause the `backend` or `backend_stream` endpoints to be slow to respond because the work is so `CPU` intensive that the container can't accept new connections or, in the event the worker crashes the system can keep accepting new commands and pick up where it left off.

## Stores

We have a number of stores in `grid` for different use cases.

### Redis

Since every object in `syft` has a `UUID` most of our data is stored in what is called `key` / `value` store. Redis is a popular high performance store which allows us to easily store or lookup any object with a particular `key` in this case a `UUID`.

Data you wish to operate on or the intermediate variables created while using `syft` usually end up in this `key`/`value` store.

If we open the `redis` view inside `TablePlus` we can see any data which gets put inside a Node's `.store` attribute.

### SeaweedFS

What is the blob storage system for you may ask. Well `redis` has a limit of `512mb` for objects to ensure it can maintain its high performance characteristics. As you can imagine real world data can often exceed those numbers by several orders of magnitute. Additionally, the way we access the `redis` store through the `backend` container, `backend_stream` container and `celeryworker` introduces a number of steps which will slow down processing of large amounts of data and require huge jumps in ram utilization.

Instead when dealing with large amounts of data, we prefer to use a seperate system called `SeaweedFS` which allows high throughput multi-threaded transfers of data between the `client` and where-ever the `seaweedfs` container is running.

By utilizing the standardized `Amazon S3` protocols we can obtain a `pre-signed` url from the `backend` for both `upload` and `download` and then allow the `client` and blob storage to interact directly.

To make this seamless to the rest of the application, we store a `ProxyData` object inside the `redis` store which contains only the meta data needed to retrieve the real object when we need it.

Normally this happens automatically for certain types of objects or sizes. One of those types is our built-in `Tensor`.
