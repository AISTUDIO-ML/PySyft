```python
def test_bnc():
    shape = (10, 3, 50, 50)

    smol_data = PhiTensor(
        child=np.random.rand(*shape)*255, 
        data_subjects=np.zeros(shape), min_vals=0, max_vals=255)

    c = nn.Convolution(3, 3, input_shape=shape)
    c.connect_to()
    bn = nn.BatchNorm()
    bn.connect_to(c)
    c_out = c.forward(smol_data)
    return bn.forward(c_out)
```

```python
def test_c_bn_avg():
    shape = (10, 3, 6, 6)

    smol_data = PhiTensor(
        child=np.random.rand(*shape)*255, 
        data_subjects=np.zeros(shape), min_vals=0, max_vals=255)

    c = nn.Convolution(3, 3, input_shape=shape)
    c.connect_to()
    bn = nn.BatchNorm()
    bn.connect_to(c)
    c_out = c.forward(smol_data)
    bn_out = bn.forward(c_out)
    avg = nn.AvgPool((2,2))
    avg.connect_to(bn)
    return avg.forward(bn_out)
    
test_c_bn_avg()
```

```python
def test_c_bn_max():
    shape = (10, 3, 6, 6)

    smol_data = PhiTensor(
        child=np.random.rand(*shape)*255, 
        data_subjects=np.zeros(shape), min_vals=0, max_vals=255)

    c = nn.Convolution(3, 3, input_shape=shape)
    c.connect_to()
    bn = nn.BatchNorm()
    bn.connect_to(c)
    c_out = c.forward(smol_data)
    bn_out = bn.forward(c_out)
    maxp = nn.MaxPool((2,2))
    maxp.connect_to(bn)
    return maxp.forward(bn_out)
    
test_c_bn_max()
```

```python
def test_lin():
    shape = (10, 3, 6, 6)

    smol_data = PhiTensor(
        child=np.random.rand(*shape)*255, 
        data_subjects=np.zeros(shape), min_vals=0, max_vals=255)
    
    lin = nn.Linear(n_out=2)
    lin.n_in = 6
    lin.connect_to()
    return lin.forward(smol_data)
    
test_lin()
```
